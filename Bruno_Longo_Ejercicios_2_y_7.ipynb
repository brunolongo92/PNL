# Ejercicio 2

```python
URL = "https://tableros.yvera.tur.ar/"
# Sin el verify = false falla la certificación
page = requests.get(URL, verify=False)

# Parseo
soup = BeautifulSoup(page.content, 'html.parser')

# buscar links
links = soup.find_all('a')

# buscar "tablero_ODS"
tablero_ODS_link = None
for link in links:
    if 'href' in link.attrs and 'tablero_ODS' in link['href']:
        tablero_ODS_link = link['href']
        break

if tablero_ODS_link:
    print("Found the link:", tablero_ODS_link)

    # Abrir link
    tablero_ODS_page = requests.get(tablero_ODS_link, verify=False)

    # Parseo
    soup = BeautifulSoup(tablero_ODS_page.content, 'html.parser')

    # Buscar parrafos
    paragraphs = soup.find_all('p')

    # Mostrar parrafos con text y strip() sin espacios ni contenedores html
    for paragraph in paragraphs:
       print(paragraph.text.strip())
```

```python
# Abre el archivo en modo de lectura ('r')
with open('Loco.txt', 'r') as archivo:
    # Lee todo el contenido del archivo
    contenido = archivo.read()
```

```python
# Expresión regular para buscar "-" o números precedidos por "!"
patron = r'\-|\!\d+'

# Función de reemplazo que retorna una cadena vacía para cada coincidencia
texto_modificado = re.sub(patron, '', contenido)

print(texto_modificado)
```

Vemos que existen guiones medios y !\d+ representando las páginas, no parecen necesarios.

```python
# Expresión regular para buscar "-" o números precedidos por "!"
patron = r'\.'

# Función de reemplazo que retorna un salto de linea por punto
texto_corregido = re.sub(patron, '\n', texto_modificado)

# Imprimir el texto modificado
print(texto_corregido)
```

Buscamos remover el exceso de espacios en blanco.

```python
# Eliminar espacios en blanco adicionales entre párrafos
texto_final = re.sub(r'\n\s*\n', '\n', texto_corregido)

# Imprimir el texto modificado sin espacios en blanco adicionales
print(texto_final)
```
